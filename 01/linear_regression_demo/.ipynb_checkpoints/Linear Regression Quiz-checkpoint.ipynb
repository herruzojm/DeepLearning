```python
import pandas as pd
from sklearn import linear_model

# Assign the dataframe to this variable.
bmi_life_data = pd.read_csv('bmi_and_life_expectancy.csv') 

# Make and fit the linear regression model
life_expectancy_values = bmi_life_data[['Life expectancy']]
bmi_values = bmi_life_data[['BMI']]

bmi_life_model = linear_model.LinearRegression()
bmi_life_model.fit(bmi_values, life_expectancy_values)

# Mak a prediction using the model
laos_life_exp = bmi_life_model.predict(21.07931)
print(laos_life_exp)

try:
    import matplotlib.pyplot as plt    
    plt.scatter(bmi_values, life_expectancy_values, color='red')
    plt.plot(bmi_values, bmi_life_model.predict(bmi_values), color='blue', linewidth=3)
    plt.show()
except:
    pass
```
